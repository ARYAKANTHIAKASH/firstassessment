```python
import requests
from bs4 import BeautifulSoup as bs

class Scraper:
    def __init__(self,url,item):
        #https://www.flipkart.com/search?q=mobiles
        self.final_url=url+'/search?q='+item
        self.pull_data(self.final_url)
        
    def pull_data(self,finalurl):
        print(finalurl)
        print("pull data function")
        
        self.result=requests.get(finalurl)
        print(self.result.status_code)
        print(self.result.text)
        self.soup=bs(self.result.content)
        print(self.soup.prettify())
        
        self.brand=self.soup.find('span',attrs={'class','B_NuCI'})
        print(self.brand)
        
        self.price=self.soup.find('div',attrs={'class','_30jeq3 _16Jk6d'})
        print(self.price)
        
        self.rating=self.soup.find('div',attrs={'class','_3LWZlK'})
        print(self.rating)
        
        self.specifications=self.soup.find('div',attrs={'class','_1YokD2 _2GoDe3'})
        print(self.specifications)
        
        self.offers=self.soup.find('div',attrs={'class','rd9nIL'})
        print(self.offers)
        
        self.everything=self.soup.findAll('div',attrs={'class','_10UF8M _3LsR0e'})
        print(self.everything)
        
        count=0
        for item in self.everything:
            count=count+1
            print(item)
            print('----'*30)
        else:
            print('Total items in 1 page={}'.format(count))
        
        
```

```python
scraper=Scraper("https://www.flipkart.com","mobiles")
```
